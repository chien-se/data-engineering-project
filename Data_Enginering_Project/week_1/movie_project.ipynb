```python
df = pd.read_csv("movie.csv")
```

```python
print("🔍 Data Overview:")
print(df)
```

```python
### Step 2: Load & Explore Data
print("🔍 Data Overview:")
print(df)

print("\n📊 Data Info:")
print(df.info())

print("\n📈 Statistics:")
print(df.describe())
```

```python
### Step 3: Data Validation & Cleaning
print("\n❌ Missing values:")
print(df.isnull().sum())

print("\n📋 Data types:")
print(df.dtypes)

# Validate values
print("\n🔎 Validation:")
print(f"Years range: {df['year'].min()} - {df['year'].max()}")
print(f"Ratings range: {df['rating'].min()} - {df['rating'].max()}")
print(f"Budget range: ${df['budget'].min()}M - ${df['budget'].max()}M")

# Fill missing data in df["rating"] with the median 
median_rating = df["rating"].median() 
df['rating'].fillna(median_rating, inplace=True)
print(f"\n✅ Filled missing ratings with median: {median_rating}")

# Check again
print("\nAfter cleaning - Missing values:")
print(df.isnull().sum())
```

```python
### Step 4: Transform & Calculate Metrics

# Calculate profit
df['profit'] = df['revenue'] - df['budget']
df['profit_margin'] = (df['profit'] / df['budget'] * 100).round(2)
df['roi'] = (df['profit'] / df['budget']).round(2)

def rating_category(rating):
    if rating >= 9:
        return "Masterpiece"
    elif rating >= 8:
        return "Excellent"
    elif rating >= 7:
        return "Good"
    else:
        return "Average"

df['ranking_category'] = df['rating'].apply(rating_category) # also work for lambda 

print("\n💰 With Metrics:")
print(df[['title', 'year', 'rating', 'profit', 'roi']])
```

```python
# Step 5: Analysis and Insight 
# What's the most profitable movie?
                                                #   df.loc[3] -> grab the entire row at index 3
most_profitable = df.loc[df['profit'].idxmax()] #  df['profit'].idxmax() -> find the index number of the highest profit row   
print(f"\n🏆 Most Profitable:")                 # e.g., returns 3
print(f"{most_profitable['title']} (${most_profitable['profit']:.1f}M profit)")

# Average metrics
print(f"\n📊 Average Metrics:")
print(f"Avg Rating: {df['rating'].mean():.2f}/10") # : is seperator .2f = show 2 decimal place -> value : .format
print(f"Avg Budget: ${df['budget'].mean():.1f}M")
print(f"Avg Revenue: ${df['revenue'].mean():.1f}M")
print(f"Avg Profit: ${df['profit'].mean():.1f}M")

# By decade
df['decade'] = (df['year'] // 10 * 10).astype(str) + "s" # astype = convert a column to diff type
print(f"\n📅 By Decade:")
print(df.groupby('decade')['rating'].mean().round(2)) # splits data into group then do something
                                                    # split movie into decade then counting each piles to fidn avg rating
                                                    # round to 2 decimal place
# Best rated
print(f"\n⭐ Top 3 Rated Movies:")
top_rated = df.nlargest(3, 'rating')[['title', 'year', 'rating']] # return the top N ROW by a column -> top 3 movie by rating 
print(top_rated)
```

```python
# Step 6: Export the result
# Saved cleaned data
df.to_csv("Movie_Cleaned.csv", index = False )
print("\n✅ Saved: movies_cleaned.csv")

# Export to JSON
# orient -> each row become one object in a dict
# indent = 2 make sure it not all in one line 
movies_json = df.to_json('movies.json', orient='records', indent=2)
print("✅ Saved: movies.json")
```
